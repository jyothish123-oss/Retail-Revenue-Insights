```python
#Task 1 (Load and Understand the Dataset)
import pandas as pd

# Load the dataset (replace 'sales.csv' with your actual filename)
df = pd.read_csv('SuperMarket Analysis.csv')

# Display first few rows
print("ðŸ”¹ First 5 rows of the dataset:")
print(df.head())

# Display data types
print("\nðŸ”¹ Data types of each column:")
print(df.dtypes)
```

```python
#Task 2(Check for missing values)
print("\nðŸ”¹ Missing values in each column:")
print(df.isnull().sum())

# If any missing values found, you can choose to drop or fill them
# Example: df.dropna(inplace=True)
# Or: df.fillna(method='ffill', inplace=True)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nðŸ”¹ Number of duplicate rows: {duplicates}")

# Remove duplicates if any
df.drop_duplicates(inplace=True)
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# If 'Time' is a string, you can convert it too
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.time

# Create new column 'Hour' for analysis (optional)
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour

# Confirm changes
print("\nðŸ”¹ Data types after conversion:")
print(df.dtypes[['Date', 'Time']])
```

```python
# Detect outliers in a column (example: Sales)
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers

# Example for 'Sales'
sales_outliers = detect_outliers_iqr(df, 'Sales')
print(f"ðŸ”¹ Number of outliers in 'Sales': {len(sales_outliers)}")

# Display top few outliers
print(sales_outliers[['Invoice ID', 'Product line', 'Sales']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical features to check
num_features = ['Sales', 'Quantity', 'Unit price']

for feature in num_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, y=feature, color='salmon')
    plt.title(f'Boxplot of {feature}')
    plt.tight_layout()
    plt.show()
```

```python
# Remove outliers from 'Sales' (if needed)
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)]
```

```python
#Check Shape and Compare Counts
print(f"ðŸ”¹ Original dataset shape: {df.shape}")
print(f"ðŸ”¹ Dataset shape without Sales outliers: {df_no_outliers.shape}")
print(f"ðŸ§® Number of rows removed: {df.shape[0] - df_no_outliers.shape[0]}")
```

```python
#Re-run Summary Stats Without Outliers
print("\nðŸ”¹ Summary statistics of 'Sales' without outliers:")
print(df_no_outliers['Sales'].describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Plot Boxplot to Confirm Outliers Are Gone
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_no_outliers, y='Sales', color='mediumseagreen')
plt.title('Boxplot of Sales (Outliers Removed)')
plt.tight_layout()
plt.show()
```

```python
#Re-ploting Sales Trends (Line Plot)
sales_cleaned = df_no_outliers.groupby('Date')['Sales'].sum()
plt.figure(figsize=(12,6))
sales_cleaned.plot(marker='o', color='teal')
plt.title('Sales Over Time (Outliers Removed)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Task 3(Data Analysis Using NumPy and Pandas)
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("SuperMarket Analysis.csv")  # Replace with actual filename if not loaded

# Convert 'Date' and 'Time'
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.time

# ðŸ‘‰ Total and Average Sales
total_sales = df['Sales'].sum()
avg_sales = df['Sales'].mean()
print(f"ðŸ”¹ Total Sales: {total_sales:.2f}")
print(f"ðŸ”¹ Average Sales: {avg_sales:.2f}")

# ðŸ‘‰ Group by Product Line
print("\nðŸ”¹ Total Sales by Product Line:")
print(df.groupby('Product line')['Sales'].sum().sort_values(ascending=False))

# ðŸ‘‰ Group by Gender
print("\nðŸ”¹ Total Sales by Gender:")
print(df.groupby('Gender')['Sales'].sum())

# ðŸ‘‰ Group by Payment Method
print("\nðŸ”¹ Total Sales by Payment Method:")
print(df.groupby('Payment')['Sales'].sum())
```

```python
#Task 4(Data Visualization Using Matplotlib and Seaborn)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#A. Bar Plot: Total Sales by Product Line
df.groupby('Product line')['Sales'].sum().sort_values().plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

```python
#B. Pie Chart: Payment Method Distribution
payment_counts = df['Payment'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

```python
#C. Box Plot: Gender vs Total Sales
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Gender', y='Sales', hue='Gender', palette='coolwarm', legend=False)
plt.title('Sales Distribution by Gender')
plt.tight_layout()
plt.show()
```

```python
#D. Heatmap: Correlation Between Numerical Features
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
#E.Line Plot: Sales Over Time
# Group sales by date
sales_over_time = df.groupby('Date')['Sales'].sum()

plt.figure(figsize=(12,6))
sales_over_time.plot(kind='line', marker='o', color='green')
plt.title('Daily Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#F.Scatter Plot: Unit Price vs Sales
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Unit price', y='Sales', hue='Product line', palette='tab10')
plt.title('Unit Price vs Sales by Product Line')
plt.xlabel('Unit Price')
plt.ylabel('Sales')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
#G.Scatter Plot: Quantity vs Total Sales
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Quantity', y='Sales', color='orange')
plt.title('Quantity vs Total Sales')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
